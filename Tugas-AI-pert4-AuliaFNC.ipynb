```python
# Nama: Aulia Fitri Nur Cahyati
# NIM: 20220801148
print("Nama: Aulia Fitri Nur Cahyati")
print("NIM: 20220801148")


# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the data manually (shortened for example)
data = {
    'Number': [1, 2, 3, 4, 5],
    'Sleep': [1.0, 2.0, 5.0, 1.0, 2.0],
    'Appetite': [1.0, 5.0, 2.0, 1.0, 5.0],
    'Interest': [1.0, 5.0, 2.0, 1.0, 5.0],
    'Fatigue': [5.0, 1.0, 2.0, 5.0, 1.0],
    'Worthlessness': [5.0, 1.0, 2.0, 5.0, 1.0],
    'Concentration': [1.0, 5.0, 2.0, 1.0, 5.0],
    'Agitation': [5.0, 1.0, 2.0, 5.0, 1.0],
    'Suicidal Ideation': [5.0, 1.0, 2.0, 5.0, 1.0],
    'Sleep Disturbance': [1.0, 5.0, 2.0, 1.0, 5.0],
    'Aggression': [5.0, 1.0, 2.0, 5.0, 1.0],
    'Panic Attacks': [5.0, 1.0, 2.0, 5.0, 1.0],
    'Hopelessness': [5.0, 1.0, 2.0, 5.0, 1.0],
    'Restlessness': [5.0, 1.0, 2.0, 5.0, 1.0],
    'Low Energy': [5.0, 1.0, 2.0, 5.0, 1.0],
    'Depression State': ['Mild', 'Moderate', 'Severe', 'No depression', 'Moderate']
}

# Create a DataFrame
df = pd.DataFrame(data)

# Display the first few rows
print("First few rows of the manually created dataset:")
print(df.head())

# Exclude the non-numeric column ("Depression State") for correlation
numeric_df = df.drop(columns=['Depression State'])

# Example of data visualization: Plotting correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Depression-Related Factors")
plt.show()

# Example: Plotting the distribution of Depression States
plt.figure(figsize=(8, 6))
sns.countplot(x='Depression State', data=df)
plt.title("Distribution of Depression States")
plt.xlabel("Depression State")
plt.ylabel("Count")
plt.show()

# Basic statistical summary
print("\nBasic statistical summary of the data:")
print(df.describe())
```
